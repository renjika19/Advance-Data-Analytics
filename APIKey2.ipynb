```python
#list

pRISBN10 = [] # primary ISBN number-10
pRISBN13 = [] # primary ISBN number-13
titlels = []  # title of book
authorls = [] # author of book
rankls = []   # rank of the book 
LWKls = []    # rank of position of book from prior week
NWls = []     # number of weeks on the list
pricels = []  # price of the book
descrls = []  # decription of the book
publls = []   # publisher of the book
bookils = []  # book image
```

```python
end = time.time() + 3 


while time.time() <= end:  
    for news in NewsPaper:
        
        try: ISBN10 = news['primary_isbn10']
        except: ISBN10 = None
            
        try: ISBN13 = news['primary_isbn13']
        except: ISBN13 = None
            
        try: title = news['title']
        except: title = None
            
        try: author = news['author']
        except: author = None
            
        try: rank = news['rank']
        except: rank = None  
            
        try: RankLastWeek = news['rank_last_week']
        except: RankLastWeek = None
            
        try: WeeksOnList = news['weeks_on_list']
        except: WeeksOnList = None 
            
        try: price = news['price']
        except: price = None 
            
        try: description = news['description']
        except: description = None 
            
        try: publisher = news['publisher']
        except: publisher = None 
            
        try: bookImage = news['book_image']
        except: bookImage = None 
            
        
            
        


            
            
        pRISBN10.append(ISBN10)
        pRISBN13.append(ISBN13)
        titlels.append(title)
        authorls.append(author)
        rankls.append(rank)
        LWKls.append(RankLastWeek)
        NWls.append(WeeksOnList)
        pricels.append(price)
        descrls.append(description)
        publls.append(publisher)
        bookils.append(bookImage)
            
```
